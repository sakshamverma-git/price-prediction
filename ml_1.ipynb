```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
house_sizes = np.random.rand(100, 1) * 200
house_prices = 50 + 300 * house_sizes + np.random.randn(100, 1) * 20

def linear_regression(X, y):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]

    theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
    return theta

def predict(X, theta):
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    return X_b.dot(theta)

theta = linear_regression(house_sizes, house_prices)

X_new = np.array([[0], [2]])
y_predict = predict(X_new, theta)

plt.scatter(house_sizes, house_prices, color='blue', label='Data points')
plt.plot(X_new, y_predict, 'r-', label='Linear regression')
plt.xlabel('House Size (1000 sq ft)')
plt.ylabel('House Price ($1000)')
plt.title('House Price Prediction')
plt.legend()
plt.grid(True)
plt.show()

predictions = predict(house_sizes, theta)
mse = np.mean((predictions - house_prices) ** 2)
print(f'Mean Squared Error: {mse:.2f}')
print(f'Learned parameters: Intercept = {theta[0][0]:.2f}, Slope = {theta[1][0]:.2f}')
```
